```python
dataset = pd.DataFrame(index=[i for i in range(15)], columns=['1 признак', '2 признак'])
dataset['1 признак'] = x
dataset['2 признак'] = y
dataset
```

## Датасет титаник 

```python
titanik = pd.read_csv('train.csv')
```

```python
titanik.head(20)
```

Вспомним и повторим, какие бывают признаки. Для визуализации это важно


1) **Численный** - просто число из некоторого диапазона, которые можно сравнивать между собой и это будет иметь смысл

Какие признак в нашем датасете численные?


2) **Бинарный** - признак, значение которого может быть равно только одному из двух вариантов: "Да"/"Нет", 0/1

Какие признаки численные?

3) **Категориальный** - признак, значение которого может быть равно из некоторого множества значений.

Это множество может представлено в виде чисел, строк или чего-то другого.
Какие признаки категориальные?

***Важно***: категориальные признаки обычно не имеет смысла сравнивать между собой, даже если они представлены в виде чисел

### Визуализация численных данных

Гистограммы - визуализация одного признака

1) Можно менять количество разбиений - bins
2) Можно менять количество

```python
bins_array = sns.plt.hist(titanik['Age'].dropna(), bins=50, color='red')
```

```python
sns.distplot(titanik['Age'].dropna(), bins=25)
```

PairPlot - визуализация двух признаков

```python
sns.pairplot(titanik[['Age', 'Fare']].dropna())
```

```python
sns.jointplot(titanik['Age'].dropna(), titanik['Fare'].dropna())
```

```python
sns.boxplot(titanik['Age'].dropna())
```

```python
sns.boxplot(titanik['Fare'])
```

## Визуализация бинарных и категориальных признаков

```python
titanik['Sex'].value_counts()
```

Как гистограмма, только для бинарных

```python
titanik['Sex'].value_counts().plot(kind='bar')
```

Или категориальных

```python
titanik['Pclass'].value_counts().plot(kind='bar')
```

Сложнее - если мы хотим посмотреть взаимодействие двух бинарных или категориальных признаков

```python
sns.countplot(x=titanik['Pclass'], hue=titanik['Sex'])
```

## А если бинарные/категориальные и численные?

```python
sns.violinplot(x='Survived', y='Age', data=titanik)
```

Еще веселее взаимодействие двух категориальных признаков, выраженное в третьем численном признаке

```python
table = titanik.pivot_table(index='Pclass', columns='Sex', values='Fare', aggfunc=np.mean).fillna(0).applymap(float)
```

```python
sns.heatmap(table, annot=True, fmt=".1f", linewidths=.5)
```

## А теперь самое интересное - начнем анализировать данные благодаря визуализации
